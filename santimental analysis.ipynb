```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import requests

def extract_text_and_save_as_txt(dataframe, url_column_index, url_id_column_index):
    for index, row in dataframe.iterrows():
        url = row.iloc[url_column_index]  # Extract URL from DataFrame
        url_id = row.iloc[url_id_column_index]  # Extract URL_ID from DataFrame
        
        try:
            # Send a GET request to the URL
            response = requests.get(url)
            response.raise_for_status()  # Raise an exception for HTTP errors
            html_content = response.text  # Get the HTML content of the page

            # Parse HTML content using BeautifulSoup
            soup = BeautifulSoup(html_content, 'html.parser')

            # Find the article title and text using specific HTML tags or classes
            title = soup.find('title').get_text()  # Assuming the title is within <title> tags
            article_text = ''
            # Extract text from paragraphs, assuming article text is within <p> tags
            for paragraph in soup.find_all('p'):
                article_text += paragraph.get_text() + '\n'

            # Create a text file with the extracted article text
            filename = f"{url_id}.txt"
            with open(filename, 'w') as file:
                file.write(f"Title: {title}\n\n")
                file.write(article_text)
            
            print(f"Article text extracted from URL '{url}' and saved to '{filename}'")
        except Exception as e:
            # If there's an error, print a message
            print(f"Error processing URL '{url}': {e}")

# Example usage:
# Assuming your DataFrame is named 'df', with URLs in the second column (index 1)
# and URL_IDs in the third column (index 2)
# You would call the function like this:
# extract_text_and_save_as_txt(df, 1, 2)'''
```

```python
def merge_test_files(input_files, output_file):
    with open(output_file, 'w', encoding='utf-8') as outfile:
        for input_file in input_files:
            with open(input_file, 'r', encoding='latin-1') as infile:
                outfile.write(infile.read())

# Example usage:
input_files = ['StopWords_Auditor.txt', 'StopWords_Currencies.txt', 'StopWords_DatesandNumbers.txt','StopWords_GenericLong.txt','StopWords_Geographic.txt','StopWords_Generic.txt','StopWords_Names.txt']  # List of input file names
output_file = 'merged_files.txt'  # Output file name

merge_test_files(input_files, output_file)
```

```python
def txt_to_list(input_data):
    # Initialize an empty list
    lines_list = []

    # Check if the input_data is a file path
    if isinstance(input_data, str):
        # Open the text file and read its contents
        with open(input_data, 'r') as file:
            # Read each line and append it to the list
            for line in file:
                lines_list.append(line.strip())  # Remove leading/trailing whitespace and newline characters
    elif isinstance(input_data, list):
        # If input_data is already a list, simply assign it to lines_list
        lines_list = input_data
    else:
        # Raise an error for unsupported input types
        raise TypeError("Input data must be either a file path (string) or a list.")

    return lines_list

# Example usage:
file_path = 'merged_files.txt'  # Replace 'example.txt' with the path to your text file
text_list = txt_to_list(file_path)
print(text_list)
```

```python
def replace_pipe_with_comma(text_list):
    modified_list = []  # Initialize an empty list to store modified strings
    for text in text_list:
        # Replace consecutive '|' with a single '|'
        cleaned_text = '|'.join(filter(None, text.split('|')))
        # Split the cleaned text into words
        words = cleaned_text.split()
        # Replace '|' with ',' for each word
        modified_words = [word.replace('|', ',') for word in words]
        # Append modified words to the modified list
        modified_words = [word.strip(', ') for word in  modified_words if word.strip(', ')]
        modified_list.extend(modified_words)
    return modified_list

# Example usage:
text_list = text_list
modified_list = replace_pipe_with_comma(text_list)
print(modified_list)
modified_list = modified_list
```

```python
import os
from nltk.tokenize import word_tokenize

def remove_stopwords_from_file(stopwords_list, file_path):
    try:
        stopwords_lower = set(word.lower() for word in stopwords_list)
        
        with open(file_path, 'r', encoding='latin-1') as file:
            content = file.read()
        
        words = word_tokenize(content)
        filtered_words = [word.lower() for word in words if word.lower() not in stopwords_lower]
        filtered_text = ' '.join(filtered_words)
        
        with open(file_path, 'w', encoding='utf-8') as file:
            file.write(filtered_text)
            
        print(f"Stopwords removed from {file_path}")
    except Exception as e:
        print(f"Error processing file {file_path}: {e}")

def remove_stopwords_from_files(stopwords_list, folder_path):
    try:
        print("Stopwords List:")
        print(stopwords_list)
        
        for file_name in os.listdir(folder_path):
            file_path = os.path.join(folder_path, file_name)
            if file_name.endswith('.txt'):
                remove_stopwords_from_file(stopwords_list, file_path)
    except Exception as e:
        print(f"Error accessing folder {folder_path}: {e}")

# Example usage:
folder_path = r'C:\Users\HP\Downloads\sentimental analysis\test files'  # Replace 'path/to/folder' with the path to your folder containing .txt files
stopwords_list = modified_list  # Your list of stopwords

remove_stopwords_from_files(stopwords_list, folder_path)
```

```python
def count_words_in_file(file_path, word_list):
    try:
        # Initialize a dictionary to store word counts
        word_counts = {word.lower(): 0 for word in word_list}
        total_count = 0  # Initialize total count
        
        # Open the file and read its content
        with open(file_path, 'r', encoding='utf-8') as file:
            content = file.read()
        
        # Tokenize the content into words
        words = content.split()
        
        # Count occurrences of words from the word list
        for word in words:
            word_lower = word.lower()
            if word_lower in word_counts:
                word_counts[word_lower] += 1
            if word_lower in word_list:
                total_count += 1  # Increment total count for words in the word list
        
        return word_counts, total_count
    except Exception as e:
        print(f"Error processing file {file_path}: {e}")
        return None, None

# Example usage:
file_path = 'blackassign0008.txt'  # Replace with the path to your .txt file
word_list = modified_list  # List of words to count occurrences of

word_counts, total_count = count_words_in_file(file_path, word_list)
if word_counts is not None:
    print("Word Counts:")
    for word, count in word_counts.items():
        print(f"{word}: {count}")
    print(f"Total Count of List Words: {total_count}")
```

```python
import os
import pandas as pd
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize


def count_syllables(word):
    """Count syllables in a word using a simple heuristic approach"""
    vowels = 'aeiouy'
    count = 0
    last_char_was_vowel = False
    for char in word.lower():
        if char in vowels:
            if not last_char_was_vowel:
                count += 1
            last_char_was_vowel = True
        else:
            last_char_was_vowel = False
    # Adjust syllable count for words ending in 'e'
    if word.lower().endswith('e') and not word.lower().endswith('le'):
        count -= 1
    # Ensure minimum syllable count is 1
    return max(count, 1)

def extract_derived_variables(folder_path, positive_words, negative_words):
    # Initialize empty lists to store results
    filenames = []
    positive_scores = []
    negative_scores = []
    polarity_scores = []
    subjectivity_scores = []
    avg_sentence_lengths = []
    pct_complex_words = []
    fog_indices = []
    avg_words_per_sentence = []
    complex_word_counts = []
    total_word_counts = []
    syllable_counts_per_word = []
    personal_pronouns_counts = []
    avg_word_lengths = []

    personal_pronouns = {'i', 'we', 'my', 'ours', 'us'}

    for filename in os.listdir(folder_path):
        if filename.endswith('.txt'):
            file_path = os.path.join(folder_path, filename)
            with open(file_path, 'r', encoding='utf-8') as file:
                text = file.read()

            # Tokenize text into words
            words = word_tokenize(text.lower())

            # Calculate derived variables
            positive_score = sum(1 for word in words if word in positive_words)
            negative_score = -sum(1 for word in words if word in negative_words)
            polarity_score = (positive_score - negative_score) / ((positive_score + negative_score) + 0.000001)
            total_words = len(words)
            sentences = sent_tokenize(text)
            total_sentences = len(sentences)

            avg_sentence_length = total_words / total_sentences
            complex_word_count = sum(1 for word in words if len(word) > 2)
            syllable_counts_per_word_value = sum(count_syllables(word) for word in words)
            personal_pronouns_count = sum(1 for word in words if word.lower() in personal_pronouns)
            avg_word_length = sum(len(word) for word in words) / total_words

            # Append results to lists
            filenames.append(filename)
            positive_scores.append(positive_score)
            negative_scores.append(negative_score)
            polarity_scores.append(polarity_score)
            subjectivity_scores.append((positive_score + negative_score) / (total_words + 0.000001))
            avg_sentence_lengths.append(avg_sentence_length)
            pct_complex_words.append(complex_word_count / total_words)
            fog_indices.append(0.4 * (avg_sentence_length + (complex_word_count / total_words)))
            avg_words_per_sentence.append(total_words / total_sentences)
            complex_word_counts.append(complex_word_count)
            total_word_counts.append(total_words)
            syllable_counts_per_word.append(syllable_counts_per_word_value)
            personal_pronouns_counts.append(personal_pronouns_count)
            avg_word_lengths.append(avg_word_length)

    # Create DataFrame
    df = pd.DataFrame({
        'Filename': filenames,
        'Positive Score': positive_scores,
        'Negative Score': negative_scores,
        'Polarity Score': polarity_scores,
        'Subjectivity Score': subjectivity_scores,
        'Average Sentence Length': avg_sentence_lengths,
        'Percentage of Complex Words': pct_complex_words,
        'Fog Index': fog_indices,
        'Average Number of Words Per Sentence': avg_words_per_sentence,
        'Complex Word Count': complex_word_counts,
        'Word Count': total_word_counts,
        'Syllable Count Per Word': syllable_counts_per_word,
        'Personal Pronouns Count': personal_pronouns_counts,
        'Average Word Length': avg_word_lengths
    })

    # Save DataFrame to CSV
    df.to_csv('derived_variables_f.csv', index=False)
```
